Shi-Tomasi Algorithm

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

# Load image in color and convert to numpy array
img_color = Image.open(r"/content/img.jpeg").convert('RGB')
img_np = np.array(img_color)

# Convert to grayscale
gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)

# Detect corners using Shi-Tomasi
corners = cv2.goodFeaturesToTrack(gray, maxCorners=200, qualityLevel=0.01, minDistance=10)

# Convert to integer coordinates
# Use np.int_ instead of the deprecated np.int0
corners = np.int_(corners)

# Draw corners on the image
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img_np, (x, y), 2, (255, 0, 0), -1)

# Show result
plt.figure(figsize=(6, 6))
plt.imshow(img_np)
plt.title("Shi-Tomasi Corner Detection")
plt.axis("off")
plt.show()
```
